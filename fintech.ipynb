### Função que define quais dados serão usados. 

Essa função inicia, execute após definir quais dados irá usar

Caso a escolha for pelo banco de dados não é necessário inicializar as celulas acima 

```
# CÉLULA DE EXECUÇÃO DA FUNÇÃO main()

def main():
  entrar = True
  while (entrar):
    entrada = input("SISTEMA DE INFORMAÇÕES DE VALORES DE AÇÕES\nDESEJA INICIAR A ENTRADA DE VALORES OU USARÁ O BANCO DE DADOS?\n Digite 'ENTRADA' para inserir dados ou 'BANCO' para usar o banco de dados. ").upper()
    if (entrada == 'ENTRADA'):
      semana1()
      semana2()
      semana3()
      semana4()
      entrar = False
    
    if (entrada =='BANCO'):
      banco1()
      banco2()
      banco3()
      banco4()
      entrar = False

main()
```

Médias semanais

### PRIMEIRA SEMANA

```
# Primeira semana

# Cria um dataframe com os valores das ações da primeira semana
primeiraSemana = pd.DataFrame({'Ação A':acoesAsemana1,
                               'Ação B':acoesBsemana1,
                               'Ação C':acoesCsemana1,
                               'Ação D':acoesDsemana1},
                               index = ['dia 1', 'dia 2', 'dia 3', 'dia 4', 'dia 5', 'dia 6', 'dia 7'])

# Calcula a média das ações da primeira semana
medias = {'Ação A':primeiraSemana['Ação A'].mean(),
          'Ação B':primeiraSemana['Ação B'].mean(),
          'Ação C':primeiraSemana['Ação C'].mean(),
          'Ação D':primeiraSemana['Ação D'].mean()}

# Ordena o dicionário de dados das médias do menor para o maior 
medias_ordenadas = sorted(medias.items(), key=lambda x:x[1]) 

print("Médias",medias,"\n")

# Seleciona o último item após a ordenação, sendo ele o de maior valor
print('A ação com maior média na primeira semana foi:', medias_ordenadas[-1],"\n") 
primeiraSemana
```

```
# Médias mensais por ação

mediasMensal = pd.DataFrame({'Ação A':mensal['Ação A'].mean(),
          'Ação B':mensal['Ação B'].mean(),
          'Ação C':mensal['Ação C'].mean(),
          'Ação D':mensal['Ação D'].mean()},
          index= ['Média_mensal',])

print("Em destaque a ação com a maior média mensal.\n")
mediasMensal.style.highlight_max(color='blue',axis=1)
```
